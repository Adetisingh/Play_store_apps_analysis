#### **1. Description :**
It contains the details of the applications on Google Play. There are 13 features that describe a given app.The Dataset is taken from kaggle.

#### **2. Dataset Description :**
**The dataset contains the following 13 columns:**
- App: The name of the App
- Category: The category of the App
- Rating: The rating of the App
- Review: The number of reviews of the App
- Size: The size of the App
- Install: The number of installs of the App
- Type: The type of the App(Free or Paid)
- Price: The price of the App
- Content Rating : The audience of the app
- Genres: The genre of the app
- Last Updated : The date when the app was last updated
- Current Ver : The current version of the app
- Android Ver : Android version required to use the app

```python
#   Replacing free with 0 and paid with 1 in Type column
df.Type.value_counts()
df['Type'] = df['Type'].str.replace('Free','0')
df['Type'] = df['Type'].str.replace('Paid','1')
```

```python
df.Type.unique()
```

```python
# Drop rows where 'Type' column has NaN values
df.drop(df.index[df['Type'].isna()], inplace=True)
```

```python
df['Type'] = df['Type'].astype(int)
```

```python
df['Type'].dtypes
```

```python
#Formatting the Last Updated column (Time Series)
df['Last Updated'] = pd.to_datetime(df['Last Updated'])
df['Last Updated'].head(5)
```

```python
df['Content Rating'].value_counts()
```

```python
df['Content Rating'] = df['Content Rating'].str.replace('Everyone','0')
df['Content Rating'] = df['Content Rating'].str.replace('Teen','13')
df['Content Rating'] = df['Content Rating'].str.replace('Mature 17+','17')
df['Content Rating'] = df['Content Rating'].str.replace('Everyone 10+','10')
df['Content Rating'] = df['Content Rating'].str.replace('Adults only 18+','18')
df['Content Rating'] = df['Content Rating'].str.replace('Unrated','0')
```

```python
df['Content Rating'] = pd.to_numeric(df['Content Rating'], errors='coerce').astype('Int64')
```

```python
df[df['Content Rating'] == '0 10+']
```

```python
df.info()
```

```python
# The distribution of App's type
plt.figure(figsize=(10, 6))
plt.title("Count of Type of Apps")
sns.countplot(x='Type', data=df)
plt.xticks(rotation=90)
plt.show()
```

####This shows that mostly apps are free and less number of paid apps.

```python
# Content Rating Distribution of Apps
plt.figure(figsize=(10, 5))
plt.title("Content Rating Distribution of Apps")
sns.histplot(x='Content Rating', data=df, kde=True)
plt.show()
```

This tells that mostly apps are available to everyone(0 refers to everyone).And a less amount of Apps are rated to applicable to Teens.

```python
#Finding most used andriod version
android_version_counts=df['Android Ver'].value_counts()
android_version_counts.idxmax()
```

Most used Andriod version is 4.1 and more than that.

```python
#Finding minimum used andriod version
android_version_counts.idxmin()
```

```python
# The distribution of Andriod Version
plt.figure(figsize=(10, 6))
plt.title("Distribution of Andriod Version")
plt.xlabel('Android Version')
plt.ylabel('Count')
android_version_counts.plot()
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```

This visualization shows that most used andriod version is 4.1 and above.
